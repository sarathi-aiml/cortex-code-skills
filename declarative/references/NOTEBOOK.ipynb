## Basic Statistics

```python
-- Row counts
SELECT COUNT(*) AS ROW_COUNT FROM SCHEMA_NAME.TABLE_NAME
```

```python
-- Value distribution for a categorical column
SELECT CATEGORY_COLUMN, COUNT(*) AS COUNT
FROM SCHEMA_NAME.TABLE_NAME
GROUP BY CATEGORY_COLUMN
ORDER BY COUNT DESC
```

```python
-- Summary statistics for numeric columns
SELECT
    COUNT(*) AS COUNT,
    AVG(NUMERIC_COLUMN) AS AVG_VALUE,
    MIN(NUMERIC_COLUMN) AS MIN_VALUE,
    MAX(NUMERIC_COLUMN) AS MAX_VALUE,
    STDDEV(NUMERIC_COLUMN) AS STD_DEV
FROM SCHEMA_NAME.TABLE_NAME
```

## Sample Queries

Common analysis patterns:

```python
-- Aggregation by category
SELECT
    CATEGORY_COLUMN,
    COUNT(*) AS COUNT,
    SUM(AMOUNT_COLUMN) AS TOTAL,
    AVG(AMOUNT_COLUMN) AS AVERAGE
FROM SCHEMA_NAME.TABLE_NAME
GROUP BY CATEGORY_COLUMN
ORDER BY TOTAL DESC
```

```python
-- Time-based analysis (if date column exists)
SELECT
    DATE_TRUNC('month', DATE_COLUMN) AS MONTH,
    COUNT(*) AS COUNT,
    SUM(AMOUNT_COLUMN) AS TOTAL
FROM SCHEMA_NAME.TABLE_NAME
GROUP BY MONTH
ORDER BY MONTH
```

```python
-- Filter and sort
SELECT *
FROM SCHEMA_NAME.TABLE_NAME
WHERE COLUMN_NAME = 'value'
ORDER BY SORT_COLUMN DESC
LIMIT 100
```
